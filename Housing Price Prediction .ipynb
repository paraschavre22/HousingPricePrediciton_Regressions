```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
from matplotlib.ticker import ScalarFormatter
import seaborn as sns
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor, OLSInfluence
from statsmodels.stats.diagnostic import het_breuschpagan, het_white
from statsmodels.graphics.regressionplots import plot_leverage_resid2
from statsmodels.graphics.gofplots import qqplot
from scipy.stats import shapiro
from sklearn.linear_model import Lasso, Ridge
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score

%config InteractiveShell.ast_node_interactivity = 'all'
%matplotlib inline
```

```python
# Map yes/no boolean fields to 1/0

housing['mainroad'] = housing['mainroad'].map({'yes': 1, 'no': 0})
housing['guestroom'] = housing['guestroom'].map({'yes': 1, 'no': 0})
housing['basement'] = housing['basement'].map({'yes': 1, 'no': 0})
housing['hotwaterheating'] = housing['hotwaterheating'].map({'yes': 1, 'no': 0})
housing['airconditioning'] = housing['airconditioning'].map({'yes': 1, 'no': 0})
housing['prefarea'] = housing['prefarea'].map({'yes': 1, 'no': 0})
housing['furnishingstatus'] = housing['furnishingstatus'].map({'furnished': 1, 'semi-furnished': 0.5, 'unfurnished' : 0})
print(housing.head())
```

# Fitting and Testing Model(OLS)

```python
features = ['area', 'bathrooms', 'airconditioning', 'stories', 'parking', 'bedrooms', 'prefarea', 'furnishingstatus', 'mainroad',
            'guestroom', 'basement', 'hotwaterheating']

X = housing[features]
y = housing['price']

# Add a constant term
X = sm.add_constant(X)

# fit the OLS model
model = sm.OLS(y, X).fit()

# Print the model summary 
print(model.summary())
```

```python
#Two things can be noted to be addressed later.

#Bedrooms has a high p-value and should be eliminated from the model.
#The condition number might suggest that there are multicollinearity problems with the model.
```

```python
#Multicollinearity occurs when predictor variables in a regression model are highly correlated with each other. 
#It can cause issues in the interpretation of regression coefficients and lead to unstable estimates. 
#One common way to assess multicollinearity is to examine the variance inflation factor (VIF) for each predictor variable.
# Calculate VIF(Variance Inflation factor)

vif_data = pd.DataFrame()
vif_data['Variable'] = X.columns
vif_data['VIF'] = [variance_inflation_factor(X.values, i)
                  for i in range(X.shape[1])]
print(vif_data)
```

```python
#Since none of the VIF values exceed 5 and all are relatively close to 1, 
#we can conclude that multicollinearity is not a major concern in this model.
```

```python
# Obtain the residuals
residuals = model.resid

# Perform the Breusch-Pagan test
bp_test = het_breuschpagan(residuals, X)
print("Breusch-Pagan test p-value:", bp_test[1])

# Plot residuals against predicted values
plt.scatter(model.fittedvalues, residuals)
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.title('Residuals vs. Predicted Values')
plt.show()
```

```python
# Perform the White test
white_test = het_white(residuals, X)
print("White test p-value:", white_test[1])
```

```python
#The p-value is extremely close to 0, which means the White's Test confirms the suspected results of the Breusch-Pagan Test. We can reject the null hypothese and conclude that heteroskedasticity exists in the model.

#We need to try to use another model to account for the spread of the residuals. With so many categorical variables, we will most likely not be able to use a Weighted Least Squares model. Before selecting a new model, I should check for outliers that are affecting the model adversely.
```

```python
# Leverage and influence plot
plot_leverage_resid2(model)
```

```python
# Cook's distance Plot

influence = OLSInfluence(model)
cooks_distance = influence.cooks_distance
plt.stem(np.arange(len(cooks_distance[0])), cooks_distance[0], markerfmt= ',' ,basefmt =' ', linefmt= '-')
plt.xlabel('Observation Index')
plt.ylabel("Cook's Distance")
plt.title("Cook's Distance Plot")
plt.show()

print(model.get_influence().summary_frame()['student_resid'])
```

# Fitting and Testing the Second Model (OLS)

```python
#remove points above 0.15 Cook's Distance, roughly resonable starting alpha for outlier removal can be 0.15;
```

```python
# Use the influence_test method to get the cook's distance
influence = model.get_influence()
cooks_distance , _ = influence.cooks_distance

# Identify indices with Cook's distance above the threshold
high_cooks_indices = cooks_distance > 0.015

#Remove observations with high Cook's distance 
filtered_housing = housing[~high_cooks_indices]

# Fit the new OLS model on the filtered dataset without bedrooms
featured_filtered = ['area', 'bathrooms', 'airconditioning', 'stories', 'parking', 'prefarea', 'furnishingstatus', 'mainroad',\
                     'guestroom', 'basement', 'hotwaterheating']
X_filtered = filtered_housing[featured_filtered] 
y_filtered = filtered_housing['price']

# Add a constatnt term
X_filtered = sm.add_constant(X_filtered)

# Fit the new OLS model on the filtered dataset
model_filtered = sm.OLS(y_filtered, X_filtered).fit()

# Print the summary of the new model
print(model_filtered.summary())

```

```python
#Several predictors remain statistically significant in the new model,
#as indicated by their low p-values (p < 0.05). These predictors include 
#"area", "bathrooms", "airconditioning", "stories", "parking", "prefarea", "furnishingstatus", "mainroad", "guestroom", "basement", 
#and "hotwaterheating". This suggests that these variables have a significant impact on housing prices after accounting for 
#influential observations.
```

```python
# Extract the independent variable from the model 
X_vif2 = X_filtered.drop(columns = 'const')

# Calculate VIF for each variable
vif_data2 = pd.DataFrame()
vif_data2["Variable"] = X_vif2.columns
vif_data2["VIF"] = [variance_inflation_factor(X_vif2.values, i)
                   for i in range(X_vif2.shape[1])]

# Display the VIF dataframe
print(vif_data2)
```

```python
#Some predictor variables exhibit high VIF values, suggesting potential
#multicollinearity issues. Specifically, "area", "bathrooms", "stories", 
#and "mainroad" have VIF values greater than 5, indicating that these variables 
#may be highly correlated with other predictors in the model.
```

```python
# Create new dataset with bathroom * stories interaction term and removing original bathrooms and stories variables; also removing mainroad
```

```python
filtered_housing2 = filtered_housing.copy()
filtered_housing2['interaction_stories_bathrooms'] = filtered_housing2['stories'] * filtered_housing2['bathrooms']
filtered_housing2.drop(['bathrooms', 'stories', 'mainroad'], axis=1, inplace=True)
filtered_housing2.dtypes
filtered_housing2.head()
```

```python
# Define the selected columns, removing 
selected_columns = ['area', 'airconditioning', 'parking', 'prefarea', 'furnishingstatus',\
                    'guestroom', 'basement', 'hotwaterheating', 'interaction_stories_bathrooms']

# Create dependent and independent variables
X_filtered2 = filtered_housing2[selected_columns]
y_filtered2 = filtered_housing2['price']

# Add a constant term before fitting the new model
X_filtered2 = sm.add_constant(X_filtered2)

# Fit the new OLS model on the filtered dataset
model_filtered2 = sm.OLS(y_filtered2, X_filtered2).fit()

# Print the summary of the new model
print(model_filtered2.summary())

# Extract the independent variable from the variable
X_vif3 = X_filtered2.drop(columns = 'const')

# Calculate VIF for each variable
vif_data3 = pd.DataFrame()
vif_data3['Variable'] = X_vif3.columns
vif_data3['VIF'] = [variance_inflation_factor(X_vif3.values, i)
                   for i in range(X_vif3.shape[1])]

print(vif_data3)
```

```python
# Perform Breusch-Pagan and White's Test.

# Obtain the residuals
residuals2 = model_filtered2.resid

# Perform the Breusch-Pagan test
bp_test2 = het_breuschpagan(residuals2, X_filtered2)
print("Breusch-Pagan test p-value:", bp_test2[1])

# Plot residuals against predicted values
plt.scatter(model_filtered2.fittedvalues, residuals2)
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.title('Residuals vs. Predicted Values of Newest Model')
plt.show()

# Perform the White test
white_test = het_white(residuals2, X_filtered2)
print("White test p-value:", white_test[1])
```

```python
#While the condition number stayed nearly exactly the same, 
#all VIF numbers are now under 5, P-values are 0, and R-squared is .712. 
#The model is currently performing reasonably well.
#To further test, the Breusch-Pagan test and White's test confirmed further heteroscedasticity, 
#which prompted me to begin checking normalization in the model with a Q-Q Plot.
```

```python
#Checking for normalization with a Q-Q plot

qqplot(residuals2, line='s')  # 's' indicates a standardized line
plt.title('Q-Q Plot of Residuals')
```

```python
# Shapiro-Wilk test for normality
statistic, p_value = shapiro(residuals2)

print(f"Shapiro-Wilk Test Statistic: {statistic}")
print(f"P-value: {p_value}")

# Interpret the results
alpha = 0.05
if p_value > alpha:
    print("The residuals appear to be normally distributed (fail to reject the null hypothesis).")
else:
    print("The residuals do not appear to be normally distributed (reject the null hypothesis).")
```

```python
# With the Q-Q Plot lines deviating from the normalization line, 
# it appears to be non-normalized values at the top of the model. 
# The Shapiro-Wilk test confirmed further non-normalization of the variables, 
# Let's identify them with a histogram.
```

# Final Feature Engineering and Last OLS Trial

```python
filtered_housing2.hist(bins=20, figsize=(15, 12))
plt.suptitle("Histograms of Variables", y=1.02, fontsize=16)
plt.tight_layout()
plt.show()
```

```python
#The variables, "price" and "area" are skewed to the left, 
# Do a logarithmic transformation on each variable to more normally distribute them.
```

```python
# Create a copy of the original dataframe
filtered_housing3 = filtered_housing2.copy()

# Apply the square root transformation to the 'price' and 'area' columns
filtered_housing3['log_price'] = np.log(filtered_housing3['price'])
filtered_housing3['log_area'] = np.log(filtered_housing3['area'])

# Drop the 'price' and 'area' columns in-place
filtered_housing3.drop(columns=['price', 'area'], inplace=True)

# Look at transformed variable
filtered_housing3.hist(bins=20, figsize=(15, 12))
plt.suptitle("Histograms of Variables", y=1.02, fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# Define the selected columns, removing 
selected_columns2 = ['log_area', 'airconditioning', 'parking', 'prefarea', 'furnishingstatus',\
                    'guestroom', 'basement', 'hotwaterheating', 'interaction_stories_bathrooms']

# Create dependent and independent variables
X_filtered3 = filtered_housing3[selected_columns2]
y_filtered3 = filtered_housing3['log_price']

# Add a constant term before fitting the new model
X_filtered3 = sm.add_constant(X_filtered3)

# Fit the new OLS model on the filtered dataset
model_filtered3 = sm.OLS(y_filtered3, X_filtered3).fit()

# Print the summary of the new model
print(model_filtered3.summary())

# Extract the independent variables from the model
X_vif4 = X_filtered3.drop(columns='const')

# Calculate VIF for each variable
vif_data4 = pd.DataFrame()
vif_data4["Variable"] = X_vif4.columns
vif_data4["VIF"] = [variance_inflation_factor(X_vif4.values, i) for i in range(X_vif4.shape[1])]

# Display the VIF dataframe
print(vif_data4)
```

```python
# Perform Breusch-Pagan and White's Test.

# Obtain the residuals
residuals3 = model_filtered3.resid

# Perform the Breusch-Pagan test
bp_test3 = het_breuschpagan(residuals3, X_filtered3)
print("Breusch-Pagan test p-value:", bp_test3[1])

# Plot residuals against predicted values
plt.scatter(model_filtered3.fittedvalues, residuals3)
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.title('Residuals vs. Predicted Values of Newest Model')
plt.show()

# Perform the White test
white_test2 = het_white(residuals3, X_filtered3)
print("White test p-value:", white_test2[1])
```

```python
# Before reversing the logarithmic transformation to make the model coefficients more interpretable,
# Check if the residuals are normally distributed.
```

```python
# Pull residuals
residuals3 = model_filtered3.resid

# Create a Q-Q plot for the residuals
qqplot(residuals3, line='s')
plt.title('Q-Q Plot of Residuals')
```

```python
# The values are not perfectly normal, but they are much closer than before the logarithmic transformation.
# Now transform back the variables and test R2 and MSE.
```

```python
log_predictions = model_filtered3.predict(X_filtered3)

# Back-transform the predictions
predictions = np.exp(log_predictions)

from sklearn.metrics import mean_squared_error, r2_score

# Assume 'y_filtered3' is your actual target data
actual_values = np.exp(y_filtered3)  # Back-transform the actual values

# Calculate metrics
mse = mean_squared_error(actual_values, predictions)
r2 = r2_score(actual_values, predictions)

print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")
```

```python
# The MSE is far too large to be able to accept the R2.
# Let's explore some of the more modern models from the Scikit Learn package. 
# Avoid trying a WLS model due to the amount of categorical variables and
# the work load it would take to establish weights.
```

# Lasso Regression

```python
# Switching to a Lasso Regression model as a Weighted Least Squares model is not viable due to the large amount of categorical variables

X_train, X_test, y_train, y_test = train_test_split(X_filtered3, y_filtered3, test_size=0.2, random_state=42)

# Standardize features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Initialize Lasso regression model
lasso_model = Lasso(alpha=0.05)  # 'alpha' is the regularization strength

# Fit the model
lasso_model.fit(X_train_scaled, y_train)

# Make predictions
y_pred = lasso_model.predict(X_test_scaled)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}")

# Display the non-zero coefficients (selected features)
selected_features = np.where(lasso_model.coef_ != 0)[0]
print(f"Selected Features: {selected_features}")

# Visualize the coefficients
plt.figure(figsize=(10, 6))
plt.bar(X_filtered3.columns, lasso_model.coef_)
plt.xlabel('Feature')
plt.xticks(rotation=45)
plt.ylabel('Coefficient Value')
plt.title('Lasso Regression Coefficients')
plt.show()
```

```python
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r_squared = r2_score(y_test, y_pred)
cross_val_scores = cross_val_score(lasso_model, X_train_scaled, y_train, cv=5, scoring='neg_mean_squared_error')
print("Cross-Validation Scores:", cross_val_scores)

print(f"Mean Squared Error: {mse}")
print(f"Root Mean Squared Error: {rmse}")
print(f"R-squared: {r_squared}")

plt.scatter(y_test, y_pred)
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual vs. Predicted Values')
# Regression line
sns.regplot(x=y_test, y=y_pred, ci=None)
plt.show()
```

# Ridge Regression

```python
# Initialize Ridge regression model

# Switching to a Lasso Regression model as a Weighted Least Squares model is not viable due to the large amount of categorical variables

X_train2, X_test2, y_train2, y_test2 = train_test_split(X_filtered3, y_filtered3, test_size=0.2, random_state=42)

# Standardize features
scaler = StandardScaler()
X_train_scaled2 = scaler.fit_transform(X_train2)
X_test_scaled2 = scaler.transform(X_test2)
# 'alpha' is the regularization strength
ridge_model = Ridge(alpha=0.05)

# Fit the model
ridge_model.fit(X_train_scaled2, y_train2)

# Make predictions
y_pred2 = ridge_model.predict(X_test_scaled2)

# Evaluate the model
mse = mean_squared_error(y_test2, y_pred2)
print(f"Mean Squared Error: {mse}")

# Display the coefficients
print(f"Coefficients: {ridge_model.coef_}")

# Visualize the coefficients
plt.figure(figsize=(10, 6))
plt.bar(X_filtered3.columns, ridge_model.coef_)
plt.xlabel('Feature')
plt.xticks(rotation=45)
plt.ylabel('Coefficient Value')
plt.title('Ridge Regression Coefficients')
plt.show()
```

```python
mse2 = mean_squared_error(y_test2, y_pred2)
rmse2 = np.sqrt(mse)
r_squared2 = r2_score(y_test2, y_pred2)

print(f"Mean Squared Error: {mse2}")
print(f"Root Mean Squared Error: {rmse2}")
print(f"R-squared: {r_squared2}")

cross_val_scores = cross_val_score(ridge_model, X_train_scaled2, y_train2, cv=5, scoring='neg_mean_squared_error')
print("Cross-Validation Scores:", cross_val_scores)

plt.scatter(y_test2, y_pred2)
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual vs Predicted Values')
# Regression line
sns.regplot(x=y_test2, y=y_pred2, ci=None)
plt.show()
```
